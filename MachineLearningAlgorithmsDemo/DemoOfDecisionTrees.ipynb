```python
def entropy(p):
    if (p==0.0 or p==1.0):
        return 0.0
    else:
        return(-p*math.log(p))
    
def gini(p):
    return(p*(1.0-p))

def find_best_split_for_a_feature(feature, df, targetColumn):
    attributeSplit= []
    classes = set(df[targetColumn])
    #print("DEBUG: Finding the best Split for " + feature )
    for value in set(df[feature]):
        #print("computing entropy for " + value)
        region = df[df[feature]==value]
        if (len(region.index) < len(df.index)):
            totalentropy = 0.0
            for classlabel in set(classes):
                examples = region[region[targetColumn]==classlabel]
                pr = (len(examples.index)+0.0)/(len(region.index)+0.0)
                #totalentropy = totalentropy + entropy(pr)
                totalentropy = totalentropy + gini(pr)
            #print(totalentropy)
            attributeSplit.append((value,totalentropy))
            #print(attributeSplit)
    if (len(attributeSplit)>0):
        attributeSplit.sort(key=lambda x:x[1])
        #print(attributeSplit)
        #print("DEBUG: Best Split for feature " + feature + " is " + attributeSplit[0][0] + " and entropy is " + str(attributeSplit[0][1]))
        return(attributeSplit[0])
    else:
        return(list(set(df[feature]))[0],1.0)
```

```python
def find_next_feature(df, features,targetColumn):
    featureSplitEntropy = []
    for feature in features:
        (attrv, e) = find_best_split_for_a_feature(feature,df,targetColumn)
        featureSplitEntropy.append((feature,attrv,e))
    featureSplitEntropy.sort(key=lambda x: x[2])
    #print(featureSplitEntropy)
    return(featureSplitEntropy[0])
```

```python
def isRegionHomogeneous(df,targetColumn):
    classes = set(df[targetColumn])
    if (len(classes)==1):
        return(True)
    return(False)

def sizeofRegion(df, threshold):
    if (len(df.index)<= threshold):
        return(True)
    return(False)
```
